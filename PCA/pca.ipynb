```python
import numpy as np
import pandas as pd
```

```python
x = pd.read_csv("../data_processed.csv").to_numpy()
```

```python
from PCA import PCA
pca = PCA(x)
```

```python
pca.explained_variance_calculation()
```

```python
n = 20
z = pca.fit_transform(n)
reconstructed_x = pca.inverse_transform(z)
reconstruction_error = np.mean((x - reconstructed_x) ** 2)
print(reconstruction_error*100)
```
