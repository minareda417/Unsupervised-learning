```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("../data_processed.csv")
y = df['diagnosis']
x = df.drop('diagnosis', axis = 1).to_numpy()
```

```python
from PCA import PCA
pca = PCA(x)
```

```python
pca.plot2d(y)
```

```python
pca.plot_covariance_matrix()
```

```python
pca.explained_variance_calculation()
```
